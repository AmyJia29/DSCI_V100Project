# Data Science Project: Planning Stage (Individual)
## Problem: Predicting Usage of a Video Game Research Server

Board Question: What player characteristics and behaviours are most predictive of subscribing to a game-related newsletter, and how do these features differ between various player types?

Research Question: **Can player's experience level, age, gender and playtime predict whether a player subscribes to the newsletter?**

Response variable: 
- subscribe (logical) from players.csv, which indicates whether the player subscribed to newsletter or not

Predictor variables: 
- experience (categorical) from players.csv, which indicates experience level of the player
- gender (categorical) from players.csv, which indicates gender of the player
- age (numeric) from players.csv, which indicates age of the player in years
- played_hours (numeric) from players.csv, which indicates total hours the player spent in the game

### Data Wrangling

```r
cat("=== DATA WRANGLING PLAN ===")
# create a clean version of the dataset for modeling
players_clean <- players |>
select(hashedEmail, subscribe, experience, Age, gender, played_hours) |>
mutate(
    subscribe = factor(subscribe, levels = c("FALSE", "TRUE")),
    experience = factor(experience, 
                        levels = c("Beginner", "Amateur", "Regular", "Veteran", "Pro"),
                        ordered = TRUE))
cat("Cleaned dataset preview:\n")
head(players_clean)
cat("\n")
cat("Cleaned dataset has", 
    nrow(players_clean), 
    ncol(players_clean), 
    "variables\n\n")
```

### Exploratory Data Visualization

#### Visualization 1: Subscription Rate by Experience Level

```r
ggplot(players, aes(x=experience, fill=subscribe))+
geom_bar(position="fill")+
labs(title="Newsletter Subscription Rate by Experience Level",
     x="Experience Level",
     y="Percentage",
     fill="Subscribed")
```

##### This stacked bar plot displays the proportion of players who subscribed compared to those who did not and is grouped by their experience level. Across all experience levels, majority of the players subscribed to the newsletter. We can see that Regular experience level has the highest subscription rates, while Veteran has the least. Amateur and Pro have very similar subscription rates.

#### Visualization 2: Subscription Rate by Age 

```r
players|>
filter(!is.na(Age))|>
ggplot(aes(x=Age, fill=subscribe))+
geom_density(alpha=0.5)+
labs(title="Age Distribution by Newsletter Subscription Status",
     x="Age (in years)",
     y="Density",
     fill="Subscribed")
```

##### This density plot compares the age distributions of subscribed and non-subscribed players. The overall shape is similar with most players in the younger age range of around the ages 15 to 25. However, we see an earlier and clearer peak for subscribers, suggesting younger players are more likely to subscribe to the newsletter. 

#### Visualization 3: Distribution of Played Hours by Subscription Status

```r
ggplot(players, aes(x=played_hours, fill=subscribe))+
geom_histogram(bins=30)+
facet_wrap(~subscribe, ncol=1, labeller=labeller(subscribe=c("TRUE"="Subscribed",
                                                             "FALSE"="Not Subscribed")))+
labs(title="Distribution of Total Hours Played by Subscription Status",
     x="Hours Played",
     y="Count")

# Summary Statistics by Subscription
players|>
group_by(subscribe)|>
summarise(mean_hours=round(mean(played_hours),2),
          median_hours=round(median(played_hours),2), 
          n=n())|>
kable(caption="Played Hours by Subscription Status")
cat("\n")
```

##### The histograms compare how many hours players have spent in total, separated by subscription status. Most players spend a small amount of time playing. Only a small group of subscribers have higher total play hours, while most play fewer than 10 hours. 

#### Visualization 4: Subscription Rate by Gender

```r
# gender summary
gender_summary <- players|>
group_by(gender, subscribe)|>
summarise(count=n(), .groups='drop')|>
group_by(gender)|>
mutate(total=sum(count),
       percentage=count/total*100)

ggplot(gender_summary, aes(x=reorder(gender, total), y=count, fill=subscribe))+
geom_col(position="fill")+
coord_flip()+
labs(title="Subscription Rate by Gender",
     x="Gender",
     y="Percentage",
     fill="Subscribed")
```

##### This stacked bar plot shows the proportion of subscribers and non-subscribers within each gender indentity category. The subscription rates are quite high for majority of the gender, except Prefer not to say, which show lower subscription rates, but this may reflect smaller group sizes rather than behavorial differences.

```r
# VISUALIZATION: AGE VS PLAYED HOURS BY SUBSCRIPTION
ggplot(players|>filter(!is.na(Age)),
       aes(x=Age, y=played_hours, color=subscribe))+
geom_point(alpha=0.6, size=3)+
labs(title="Relationship between Age and Played Hours",
     x="Age (in years)",
     y="Hours Played",
     color="Subscribed")
```

##### This scatter plot shows how total hours played relate to age and whether the player subscribes. Most players are between the ages 15 to 25 and some young subscribers between the ages 18 and 20 show extremely high amount of hours played in the game. 

#### Methods and Plan
Proposed method: K-Nearest Neighbors (K-NN) Classification
This method is appropriate because we are predicting newsletter subscription status and our response variable, which indicates whether a player has subscribed or not, is binary. And K-NN is well suited for such categorical classification tasks. There is no strong assumptions, unlike regression, K-NN classification makes no assumptions about linear relationships or variable distributions. Additionally, it can handle a mix of numeric (age, played_hours) and categorical (experience, gender) variables. 

We are going to assume that all predictor variables (experience, age, gender, and played_hours) are relevant to subscription behavior and measured on comparable scales. The dataset has sufficient number of data points in each class so that neighborhoods are meaningful. Also, the numerical and categorical variables should be properly scaled so that distance calculations reflect similarity.\

Potential limitations could be the 4 predictor variables, we may face challenges if most are included. Variables must be standardized and imbalanced data could cause predictions to be biased toward the majority class. Also, K-NN can be slow for large datasets, as it computes distances to all training sets.

We will use cross-validation instead of validation set because it provides a more reliable estimate of model performance by averaging results across multiple folds. And it maximizes data efficiency, allowing all data points in the training set to be used for both training and validation.

##### Data Processing Plan:
##### Data Cleaning:
- indentify and address missing values
- check for and handle any duplicate records
- ensure all categorical levels are appropriate and properly labeled
- convert categorical predictors into numeric form
- standardize all numeric variables to ensure fair distance comparisons
##### Data Splitting:
- split the dataset into training and testing sets
- set a random seed for reproducibility
- data will be split before any model training or cross-validation
##### Model Training and Selection:
- train K-NN models across a range of k values
- use cross-validation within the training set to identify the best k value
##### Model Evaluation:
- evaluate the final model on the test set using accuracy and precision
